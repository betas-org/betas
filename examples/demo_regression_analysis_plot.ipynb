### Setup

```python
import sys
sys.path.insert(0, '../betas')

# Display all of the results.
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

# Hide warnings
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import regression_analysis_plot as plt
```

### Load College Dataset from data

```python
# import college data
college = pd.read_csv('../data/college.csv', sep=',', header=0)
college.rename(columns={'Unnamed: 0': 'College'}, inplace=True)
college = college.set_index('College')
```

### Initialize Regression Analysis Plot Object

```python
myplot = plt.RegressionAnalysisPlot(college)
```

### Get/Set Predictor Variable(s) and Response Variable

```python
print('Currect predictor variable(s):\n', myplot.get_predictors())
print('Current response variable:\n', myplot.get_response())
```

```python
X = list(college.columns[1:4])
Y = 'Private'
myplot.set_predictors(X)
myplot.set_response(Y)
print('Currect predictor variable(s):\n', myplot.get_predictors())
print('Currect response variable:\n', myplot.get_response())
```

### Create Matrix Scatter Plot

```python
myplot.matrix_plot()
```

### Create Correlation Heatmap

```python
myplot.corr_heatmap(figsize=(15,15))
```

### Create Scatter Plot with Regression Line

```python
myplot.reg_plot('Accept', 'Enroll')
```

### Create Box Plot

```python
myplot.box_plot('Private', 'Accept')
```

### Create Distribution Plot with Category Labeled

```python
myplot.dist_plot('Grad.Rate', 'Private')
```

### Fit Linear Regression Model with Wrong Variable Type

```python
# Demo of error: Will raise data type error if choose categorical variable.
myplot.reg()
print('Predictors:', myplot.get_predictors())
print('Response:', myplot.get_response())
print('Data types:\n', college.dtypes)
```

### Reset Response Variable and Fit Linear Regression Model

```python
myplot.set_response('Expend')
myplot.reg(report=True)
```

```python
myplot.resid_plot()
```

```python
myplot.qq_plot()
```
